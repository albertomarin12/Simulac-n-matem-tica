$$ f(x) = 1260+117x-9x^2 $$

```python
import sympy as sp
```

```python
def f(x) :
    return 1260 + 117*x -9*x**2
sp.var("x")
f(x)
```

```python
df= sp.diff(f(x),x)
df
```

```python
pc= sp.solve(df,x)
pc
```
